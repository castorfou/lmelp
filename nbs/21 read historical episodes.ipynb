```python
from bs4 import BeautifulSoup

# Chemin vers le fichier HTML
file_path = "/home/guillaume/git/lmelp/db/À écouter plus tard I Radio France/À écouter plus tard I Radio France.html"

# Lire le contenu du fichier HTML
with open(file_path, "r", encoding="utf-8") as file:
    html_content = file.read()

# Analyser le contenu HTML avec BeautifulSoup
soup = BeautifulSoup(html_content, "html.parser")

# Extraire les informations des épisodes
episodes = []

# Rechercher les éléments contenant les informations des épisodes
for item in soup.find_all("li", class_="Collection-section-items-item"):
    title_element = item.find("span", class_="CardTitle")
    link_element = item.find("a", class_="underline-hover")
    description_element = item.find("div", class_="CardDescription")
    date_elements = item.find("div", class_="DefaultDetails-secondLine").find_all("p")

    if (
        title_element
        and link_element
        and description_element
        and len(date_elements) >= 3
    ):
        title = title_element.get_text(strip=True)
        url = link_element["href"]
        description = description_element.get_text(strip=True)
        date = date_elements[0].get_text(strip=True)
        duration = date_elements[2].get_text(strip=True)

        episodes.append(
            {
                "title": title,
                "url": url,
                "description": description,
                "date": date,
                "duration": duration,
            }
        )

# Afficher les informations des épisodes
for episode in episodes:
    print(f"Title: {episode['title']}")
    print(f"URL: {episode['url']}")
    print(f"Description: {episode['description']}")
    print(f"Date: {episode['date']}")
    print(f"Duration: {episode['duration']}")
    print()
```

```python
import requests
from bs4 import BeautifulSoup
import json


def get_audio_url(episode):
    """
    Prend un épisode en entrée et retourne l'URL vers le fichier audio .m4a ou .mp3.
    """
    url = episode["url"]

    try:
        # Faire une requête HTTP pour obtenir le contenu de la page
        response = requests.get(url)
        response.raise_for_status()  # Vérifier que la requête a réussi
    except requests.RequestException as e:
        print(f"Erreur lors de la requête HTTP: {e}")
        return None

    # Analyser le contenu HTML avec BeautifulSoup
    soup = BeautifulSoup(response.content, "html.parser")

    # Rechercher la balise <script> contenant l'objet JSON
    script_tag = soup.find("script", string=lambda t: t and "contentUrl" in t)

    if script_tag:
        try:
            # Extraire le contenu JSON de la balise <script>
            json_text = script_tag.string
            json_data = json.loads(json_text)

            # Extraire l'URL du fichier audio
            audio_url = None
            for item in json_data.get("@graph", []):
                if item.get("@type") == "RadioEpisode":
                    main_entity = item.get("mainEntity", {})
                    audio_url = main_entity.get("contentUrl")
                    break

            return audio_url
        except (json.JSONDecodeError, KeyError, TypeError) as e:
            print(f"Erreur lors de l'analyse du JSON: {e}")
            return None

    print("Balise <script> contenant 'contentUrl' non trouvée")
    return None


audio_url = get_audio_url(episodes[0])
print(f"Audio URL: {audio_url}")
```

```python
from datetime import datetime
import locale

locale.setlocale(locale.LC_TIME, "fr_FR.UTF-8")  # 'fr_FR' selon le système

# Dictionnaire de remplacement pour normaliser les abréviations des mois
month_replacements = {
    "janv.": "janv.",
    "fév.": "févr.",
    "mars": "mars",
    "avr.": "avril",
    "mai": "mai",
    "juin": "juin",
    "juill.": "juil.",
    "août": "août",
    "sept.": "sept.",
    "oct.": "oct.",
    "nov.": "nov.",
    "déc.": "déc.",
}


# Fonction pour normaliser les dates
def normalize_date(date_str):
    for fr_month, fr_month_norm in month_replacements.items():
        date_str = date_str.replace(fr_month, fr_month_norm)
    return date_str


# Format de la date en francais après normalisation
date_format = "%d %b %Y"

# Transformer les dates en objets datetime
datetime_dates = [
    datetime.strptime(normalize_date(date), date_format) for date in exemples_dates
]

# Afficher les objets datetime
for dt in datetime_dates:
    print(dt)
```
